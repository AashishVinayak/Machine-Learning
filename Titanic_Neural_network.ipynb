```
def one_hot(vector):
    
    vector = np.array(vector)

    # vector should be a vector not an array
    if(vector.ndim !=1):
        return false

    # length of vector
    vector_len = len(vector)

    # extract the categories
    categories = list(set(vector))

    # Total number of categories
    categories_len = len(categories)

    # Initialize a numpy array of size ( vector_len, categories_len )
    one_hot_vector = np.zeros([vector_len, categories_len], dtype=np.float)

    for i in range(vector_len):

        # for every row in the vector, check if the category matches with a value 
        # in the categories list
        for c in range(categories_len):
            # replace the zero with a one at the column index where the category matches
            if vector[i] == categories[c]:
                one_hot_vector[i, c] = 1

    return one_hot_vector
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame(pd.read_csv('TestBench/train.csv'))

# filling missing data with last valid observation
data = data.fillna(method='ffill')

# one hot encoding nominal attributes 
sex_one_hot = one_hot(data['Sex'])
data['is_male'] = sex_one_hot[:, 0]
data['is_female'] = sex_one_hot[:, 1]

embarked_one_hot = one_hot(data['Embarked'])
data['embarked_S'] = embarked_one_hot[:, 0]
data['embarked_Q'] = embarked_one_hot[:, 1]
data['embarked_C'] = embarked_one_hot[:, 2]

pclass_one_hot = one_hot(data['Pclass'])
data['pclass_1'] = embarked_one_hot[:, 0]
data['pclass_2'] = embarked_one_hot[:, 1]
data['pclass_3'] = embarked_one_hot[:, 2]
```

```
print('These are the attributes we will use\n')
feature_list = ['Pclass', 'is_male', 'is_female', 'Age', 'SibSp', 'Parch', 'Fare', 'embarked_C', 
                'embarked_S','embarked_Q', 'pclass_3','pclass_1', 'pclass_2']
data[feature_list]
```

```
# splitting data into training and validation set

# split ratio: 70:30

split_ratio = int(len(data)*0.75)

x_train = np.array(data[feature_list].iloc[0:split_ratio])
y_train = np.array(data[['Survived']].iloc[0:split_ratio])

x_test  = np.array(data[feature_list].iloc[split_ratio:])
y_test  = np.array(data[['Survived']].iloc[split_ratio:])
```

```
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(solver='adam', alpha=0.01, hidden_layer_sizes=(30,20), max_iter=2500)

model.fit(x_train, y_train)
```

```
# now calculating the training and validation accuracy
from sklearn.metrics import accuracy_score

x_train_pred = model.predict(x_train)
x_test_pred = model.predict(x_test)

print("Training accuracy: ",accuracy_score(y_train, x_train_pred))
print("Testing accuracy: ",accuracy_score(y_test, x_test_pred))
```
