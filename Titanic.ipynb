```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame(pd.read_csv('TestBench/train.csv'))

# filling missing data with last valid observation
data = data.fillna(method='ffill')

# encoding nominal attributes 
for i in range(len(data['Sex'])):
    if data['Sex'].iloc[i] == 'male':
        data['Sex'].iloc[i] = 1
    else:
        data['Sex'].iloc[i] = 0


for i in range(len(data['Embarked'])):
    if data['Embarked'].iloc[i] == 'S':
        data['Embarked'].iloc[i] = 1
    elif data['Embarked'].iloc[i] == 'C':
        data['Embarked'].iloc[i] = 2
    else:
        data['Embarked'].iloc[i] = 3
```

```
print('These are the attributes we will use\n')
data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Survived']]
```

```
# splitting data into training and validation set

# split ratio: 70:30

split_ratio = int(len(data)*0.7)

x_train = np.array(data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].iloc[0:split_ratio])
y_train = np.array(data[['Survived']].iloc[0:split_ratio])

x_test  = np.array(data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].iloc[split_ratio:])
y_test  = np.array(data[['Survived']].iloc[split_ratio:])
```

```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(penalty='l2', solver='liblinear')

model.fit(x_train, y_train)
```

```
# now calculating the training and validation accuracy
from sklearn.metrics import accuracy_score

x_train_pred = model.predict(x_train)
x_test_pred = model.predict(x_test)

print("Training accuracy: ",accuracy_score(y_train, x_train_pred))
print("Testing accuracy: ",accuracy_score(y_test, x_test_pred))
```

```
# making predictions for test data 
test_data = pd.DataFrame(pd.read_csv('TestBench/test.csv'))

# filling missing data with last valid observation
test_data = test_data.fillna(method='ffill')

# encoding nominal attributes 
for i in range(len(test_data['Sex'])):
    if test_data['Sex'].iloc[i] == 'male':
        test_data['Sex'].iloc[i] = 1
    else:
        test_data['Sex'].iloc[i] = 0


for i in range(len(test_data['Embarked'])):
    if test_data['Embarked'].iloc[i] == 'S':
        test_data['Embarked'].iloc[i] = 1
    elif test_data['Embarked'].iloc[i] == 'C':
        test_data['Embarked'].iloc[i] = 2
    else:
        test_data['Embarked'].iloc[i] = 3
```

```
test_pred = model.predict(test_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']])
```
